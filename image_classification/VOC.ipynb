```python
datasetPath = '/Users/zhanghao/Documents/BigData/project/CBIR/dataset/'
descriptorName = 'SURF'
descriptorPath = '/Users/zhanghao/Documents/BigData/project/CBIR/descriptors/datasetDescriptors'
dict={"SURF":describeSURF,"SIFT":describeSIFT,"ORB":describeORB}
className = ['train','cat','chair','sofa','tvmonitor' ]
classDatasetPath = datasetPath + 'aeroplane'
descriptors = getDescriptors(classDatasetPath, dict[descriptorName])
for name in className:
    print(name)
    classDatasetPath = datasetPath  + name
    tempDescriptors=getDescriptors(classDatasetPath, dict[descriptorName])
    descriptors = np.vstack((descriptors,tempDescriptors))
print(descriptors.size)
#writting the output
file=descriptorPath+".pickle"

with open(file, 'wb') as f:
	pickle.dump(descriptors, f)
```

# Construct a visual dictionary from the training descriptors

```python
visualDictPath = '/Users/zhanghao/Documents/BigData/project/CBIR/visualDictionary/visualDictionary'
descriptorFilePath = descriptorPath+'.pickle'
visualWordsNum = 16
print("estimating a visual dictionary of size: "+str(visualWordsNum)+ " for descriptors in path:"+descriptorPath)
with open(descriptorFilePath, 'rb') as f:
    descriptors=pickle.load(f)

visualDictionary=kMeansDictionary(descriptors,visualWordsNum)

#output
file=visualDictPath+".pickle"

with open(file, 'wb') as f:
	pickle.dump(visualDictionary, f)

print("The visual dictionary  is saved in "+file)
```

# Compute VLAD descriptors from the visual dictionary

```python
#estimating VLAD descriptors for the whole dataset
#print("estimating VLAD descriptors using "+descriptorName+ " for dataset: /"+datasetPath+ " and visual dictionary: /"+visualDictPath)
datasetPath = '/Users/zhanghao/Documents/BigData/project/CBIR/dataset'
descriptorName = 'SURF'
visualDictPath = '/Users/zhanghao/Documents/BigData/project/CBIR/visualDictionary/visualDictionary'
visualDictFilePath = visualDictPath + '.pickle'
VLADDescriptorPath = '/Users/zhanghao/Documents/BigData/project/CBIR/VLADdescriptors/VLADDescriptors'
className = ['train','cat','chair','sofa','tvmonitor']

with open(visualDictFilePath, 'rb') as f:
    visualDictionary=pickle.load(f) 
print (visualDictionary)
#computing the VLAD descriptors
dict={"SURF":describeSURF,"SIFT":describeSIFT,"ORB":describeORB}
classDatasetPath = datasetPath + '/aeroplane'
V, idImages = getVLADDescriptors(classDatasetPath,dict[descriptorName],visualDictionary)
print(len(idImages))
idClasses = np.ones(len(idImages))
classes_id = 1
for name in className:
    classes_id = classes_id + 1
    classDatasetPath = datasetPath + '/' + name
    temp_V, temp_idImages = getVLADDescriptors(classDatasetPath,dict[descriptorName],visualDictionary)
    print(len(temp_idImages))
    temp_idClasses = np.ones(len(temp_idImages))
    for i in range(len(temp_idImages)):
        temp_idClasses[i] = classes_id
    idClasses = np.hstack((idClasses,temp_idClasses))
    V = np.vstack((V,temp_V))
    idImages = idImages + temp_idImages
#output
print(len(idImages))
print(V.shape)
print(len(idClasses))
#PCA
#pca = PCA(n_components=256)
#newV = pca.fit_transform(V)
#print(newV.shape)
file=VLADDescriptorPath+".pickle"

with open(file, 'wb') as f:
	pickle.dump([idImages, V, datasetPath, idClasses], f)

print("The VLAD descriptors are  saved in "+file)
```

# Make an index from VLAD descriptors using a ball-tree DS

```python
VLADDescriptorPath = '/Users/zhanghao/Documents/BigData/project/CBIR/VLADdescriptors/VLADDescriptors'
VLADDescriptorFilePath = VLADDescriptorPath + '.pickle'
leafSize = 40
indexBallTreePath = '/Users/zhanghao/Documents/BigData/project/CBIR/ballTreeIndexes/ballTreeIndexes'
#estimating VLAD descriptors for the whole dataset
print("indexing VLAD descriptors from "+VLADDescriptorPath+ " with a ball tree:")

#load the vlad descriptors VD=[imageID, VLADdescriptors, pathToImageDataSet]
with open(VLADDescriptorFilePath, 'rb') as f:
    VLAD_DS=pickle.load(f)

imageID=VLAD_DS[0]
print(len(imageID))
V=VLAD_DS[1]
pathImageData=VLAD_DS[2]
imageClass = VLAD_DS[3]

tree = indexBallTree(V,leafSize)
#output
file=indexBallTreePath+".pickle"

with open(file, 'wb') as f:
	pickle.dump([imageID,tree,pathImageData,imageClass], f,pickle.HIGHEST_PROTOCOL)

print("The ball tree index is saved at "+file)
```
